```python
# Univariate analysis of categorical variables
def univariate_cat(data, column_name):
    print(f'Univariate analysis of {column_name}')
    sns.countplot(data[column_name])
    plt.show()

for col in ['Gender',  'Degree', 'Specialization']:
    univariate_cat(df, col)

# Univariate analysis of numerical variables
def univariate_num(data, column_name):
    print(f'Univariate analysis of {column_name}')
    sns.histplot(data[column_name], kde=True)
    plt.show()

for col in ['10percentage', '12percentage', 'collegeGPA','CollegeTier']:
    univariate_num(df, col)

# Bivariate analysis

sns.countplot(x='Gender', hue='Degree', data=df)
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Count Plot of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Gender': ['Male', 'Female', 'Male', 'Male', 'Female'],
    'College GPA': ['3.5-4.0', '3.0-3.5', '3.5-4.0', '3.0-3.5', '3.5-4.0']
}

# Create DataFrame
df = pd.DataFrame(data)

# Aggregate data to get counts of each GPA range for each gender
agg_data = df.groupby(['Gender', 'College GPA']).size().unstack(fill_value=0)

# Plot the stacked bar plot
agg_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.title('Gender vs College GPA')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability

# Show plot
plt.legend(title='College GPA')
plt.show()
```
