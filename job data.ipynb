```python
from scipy.stats import chi2_contingency

df['Gender'] = df['Gender'].astype('category')
df['Specialization'] = df['Specialization'].astype('category')
# Create contingency table
contingency_table = pd.crosstab(df['Gender'],df['Specialization'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
# Print results
print("Chi-squared Statistic:", chi2)
print("P-value:", p)
```

null hypothesis:there is no association between gender and specialization



alternative hypothesis:there is an association between gender and specialization

Observation from result
as the p value is 0.00000124 we reject null hypothesis


based on the p-value we conclude that there is a association between gender and specialization
